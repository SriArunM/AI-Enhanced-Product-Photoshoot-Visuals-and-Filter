```python
# Load models for image generation
import torch
from diffusers import StableDiffusionXLPipeline
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16
)
pipe = pipe.to("cuda")
```

```python
import os
import cv2
import gradio as gr
import numpy as np
import supervision as sv
import torch
from ultralytics import YOLO
from inference.models import YOLOWorld
import efficientvit
from efficientvit.models.efficientvit.sam import EfficientViTSamPredictor
from efficientvit.sam_model_zoo import create_sam_model
from diffusers import StableDiffusionXLPipeline

# Download model weights.
os.system("make model")

# Load models.
yolo_world = YOLOWorld(model_id="yolo_world/l")
device = "cuda" if torch.cuda.is_available() else "cpu"
sam = EfficientViTSamPredictor(
    create_sam_model(name="xl1", weight_url="xl1.pt").to(device).eval()
)
```

```python
# Load annotators.
BOUNDING_BOX_ANNOTATOR = sv.BoundingBoxAnnotator()
MASK_ANNOTATOR = sv.MaskAnnotator()
LABEL_ANNOTATOR = sv.LabelAnnotator()

# Define the list of classes to detect.
predefined_classes = [
    "Shoe", "Sneaker", "Bottle", "Cup", "Sandal", "Perfume", "Toy", "Sunglasses", "Car",
    "Water Bottle", "Chair", "Office Chair", "Can", "Cap", "Hat", "Couch", "Wristwatch",
    "Glass", "Bag", "Handbag", "Baggage", "Suitcase", "Headphones", "Jar", "Vase"
]

# Global variable to store masks and the original image.
global_masks = []
global_image = None
```

```python


# Object Detection Function.
def detect(image: np.ndarray, confidence_threshold: float, nms_threshold: float) -> np.ndarray:
    global global_masks, global_image
    global_image = image.copy()
    yolo_world.set_classes(predefined_classes)
    results = yolo_world.infer(image, confidence=confidence_threshold)
    detections = sv.Detections.from_inference(results).with_nms(class_agnostic=True, threshold=nms_threshold)

    sam.set_image(image, image_format="RGB")
    masks = []
    for xyxy in detections.xyxy:
        mask, _, _ = sam.predict(box=xyxy, multimask_output=False)
        masks.append(mask.squeeze())
    detections.mask = np.array(masks)

    global_masks = masks  # Save masks globally

    output_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    labels = [
        f"{predefined_classes[class_id]}: {confidence:.2f}"
        for class_id, confidence in zip(detections.class_id, detections.confidence)
    ]
    output_image = MASK_ANNOTATOR.annotate(output_image, detections)
    output_image = BOUNDING_BOX_ANNOTATOR.annotate(output_image, detections)
    output_image = LABEL_ANNOTATOR.annotate(output_image, detections, labels=labels)
    return cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
```

```python


# Image Enhancement Function.
def enhance_segmented(brightness: int, contrast: int, color: str) -> np.ndarray:
    global global_masks, global_image
    if global_image is None or not global_masks:
        return None

    enhanced_image = global_image.copy()
    color = tuple(int(color[i:i+2], 16) for i in (1, 3, 5))  # Convert hex color to RGB tuple

    for mask in global_masks:
        mask_indices = mask > 0
        segment = enhanced_image[mask_indices]

        # Adjust brightness and contrast
        segment = cv2.convertScaleAbs(segment, alpha=contrast/50, beta=brightness-50)

        # Apply color adjustment
        color_overlay = np.full_like(segment, color, dtype=np.uint8)
        segment = cv2.addWeighted(segment, 0.7, color_overlay, 0.3, 0)

        enhanced_image[mask_indices] = segment

    return enhanced_image
```

```python


# # Load image generation model.
# pipe = StableDiffusionXLPipeline.from_pretrained(
#     "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16
# )
# pipe = pipe.to("cuda")

# Image Generation Function.
def generate_product_image(description, lighting, background, angle, num_inference_steps, guidance_scale):
    prompt = f"A {description} photographed with {lighting}, on a {background}, from a {angle}."
    image = pipe(prompt=prompt, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale).images[0]
    return image
```

```python

# Function to filter images in a folder.
def filter_images_in_folder(files: list, confidence_threshold: float, nms_threshold: float) -> list:
    filtered_images = []
    for file in files:
        image_path = file.name
        image = cv2.imread(image_path)
        if image is not None:
            detect(image, confidence_threshold, nms_threshold)
            if len(global_masks) > 0:  # Check if any objects were detected
                filtered_images.append(image)
    return filtered_images
```

```python



# Create Gradio Tabs for switching between functionalities.
with gr.Blocks() as demo:
    with gr.Tabs():
        with gr.TabItem("Gen AI Visual"):
            with gr.Row():
                with gr.Column():
                    description = gr.Textbox(lines=2, placeholder="Enter the product description here...", label="Product Description")
                    lighting = gr.Radio(choices=[
                        "Natural light", "Studio lighting", "Directional lighting", "Diffuse lighting", "Ambient lighting"
                    ], label="Lighting Condition")
                    background = gr.Radio(choices=[
                        "Plain white background", "Colored background", "Textured background", "Lifestyle background", "Outdoor/Natural background"
                    ], label="Background")
                    angle = gr.Radio(choices=[
                        "Top view", "Front view", "Side view", "45-degree angle", "Close-up/Macro", "Perspective shot"
                    ], label="Angle")
                    num_inference_steps = gr.Slider(1, 100, value=50, label="Number of Inference Steps")
                    guidance_scale = gr.Slider(0.0, 20.0, value=7.5, label="Guidance Scale")
                    generate_button = gr.Button("Generate")
                with gr.Column():
                    output_image_gen = gr.Image(type="pil")

                generate_button.click(fn=generate_product_image, inputs=[description, lighting, background, angle, num_inference_steps, guidance_scale], outputs=output_image_gen)

        with gr.TabItem("Product Recognition Filter"):
            with gr.Row():
                with gr.Column():
                    input_image = gr.Image(type="numpy", label="Input Image")
                    confidence_slider = gr.Slider(minimum=0, maximum=1, value=0.3, step=0.01, interactive=True, label="Confidence Threshold")
                    nms_slider = gr.Slider(minimum=0, maximum=1, value=0.5, step=0.01, interactive=True, label="NMS Threshold")
                    detect_button = gr.Button("Detect")
                with gr.Column():
                    output_image = gr.Image(type="numpy", label="Output Image")

                detect_button.click(fn=detect, inputs=[input_image, confidence_slider, nms_slider], outputs=output_image)

                # Enhance Segments
                with gr.Column():
                    brightness_slider = gr.Slider(minimum=0, maximum=100, value=50, step=1, interactive=True, label="Brightness")
                    contrast_slider = gr.Slider(minimum=0, maximum=100, value=50, step=1, interactive=True, label="Contrast")
                    color_picker = gr.ColorPicker(value="#FFFFFF", label="Color")
                    enhance_button = gr.Button("Enhance Segments")
                    enhanced_output_image = gr.Image(type="numpy", label="Enhanced Image")

                enhance_button.click(fn=enhance_segmented, inputs=[brightness_slider, contrast_slider, color_picker], outputs=enhanced_output_image)


        with gr.TabItem("Batch Image Filtering"):
            with gr.Row():
                with gr.Column():
                    folder_input = gr.Files(label="Upload Folder", file_count="multiple")
                    confidence_slider_folder = gr.Slider(minimum=0, maximum=1, value=0.3, step=0.01, interactive=True, label="Confidence Threshold")
                    nms_slider_folder = gr.Slider(minimum=0, maximum=1, value=0.5, step=0.01, interactive=True, label="NMS Threshold")
                    filter_button = gr.Button("Filter Images")
                with gr.Column():
                    filtered_images_output = gr.Gallery(label="Filtered Images")

                filter_button.click(fn=filter_images_in_folder, inputs=[folder_input, confidence_slider_folder, nms_slider_folder], outputs=filtered_images_output)

# Launch the interface.
demo.launch(debug=True, share=True)
```
